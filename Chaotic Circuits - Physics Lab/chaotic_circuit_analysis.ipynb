```python
# Import packages needed
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy as sp
import scipy.integrate
import plotly.express as px
import matplotlib as mpl 
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
# Define function that loads data
def load_data(rv):
    t, dx, x = np.genfromtxt('data_17-02-2022/' + 'CSV' + str(rv) + '.csv', skip_header=12, delimiter=',', unpack=True)
    return t, dx, x
```

```python
# Define plotting functions
def time_series_plot(t, x, dx, Rv):
    plt.figure(figsize = (6, 4))
    plt.plot(t, x, '-', color = 'black', label = f"x, $R_v$ = {Rv} $K\Omega$")
    #plt.title("Time Series \nX")
    plt.ylabel('Voltage (V)')
    plt.xlabel('T (t/RC)') 
    plt.legend(loc = "lower right")
    # plt.savefig("Time series (exp), x, Rv"+str(Rv), format = 'jpeg')
    plt.show()
    
    plt.figure(figsize = (6, 4))
    plt.plot(t, dx, '-', color = 'red', label = f"x', $R_v$ = {Rv} $K\Omega$")
    #plt.title("Times Series \nX'")
    plt.ylabel('Voltage (V)')
    plt.xlabel('T (t/RC)')
    plt.legend(loc = "lower right")
    # plt.savefig("Time series (exp), x', Rv"+str(Rv), format = 'jpeg')
    plt.show()

    
def phase_diagram(x, dx, Rv):
    plt.figure(figsize = (6, 4))
    plt.plot(x, dx, '.', markersize="2", label = f'$R_v$ = {Rv} $K\Omega$')
    #plt.title("Phase Diagram")
    plt.xlabel("X")
    plt.ylabel("X'")
    plt.legend(loc = "lower right")
    # plt.savefig("Phase diagram (exp), Rv"+str(Rv), format = 'jpeg')
    plt.show()
```

### Bifurcation Diagrams

```python
# Plot bifurcation diagram for experimental data

# Make a list with all measured Rv values
Rv_val = []
for i in range(101):
    rv_value = i + 40
    Rv_val.append(rv_value)

more_rv = [54.1, 54.2, 54.3, 54.4, 54.5, 54.6, 54.7, 54.8, 54.9, 55.1, 55.2, 55.3, 55.4, 55.5, 55.6, 55.7, 55.8, 55.9, 69.1, 69.2, 69.3, 69.4, 69.5, 69.6, 69.7, 69.8, 69.9, 72.1, 72.2, 72.3, 72.4, 72.5, 72.6, 72.7, 72.8, 72.9, 150, 160, 170, 180, 185, 190, 195, 200, 205, 210, 220, 230, 240, 250]
for i in range(len(more_rv)):
    Rv_val.append(more_rv[i])

# Make lists of measured values for amplitude
rv_s = []
sol_s = []
for i in range(len(Rv_val)):
    Rv = Rv_val[i]
    t, dx, x = load_data(Rv)

    amp = []
    for n in range(len(x)):
        if (n > 70 and n < (len(x)-70)):
            prev = x[n-70]
            current = x[n]
            nextt = x[n+70]
            if (prev < current and nextt < current):
                amp.append(current)

    for m in range(len(amp)):
        rv_s.append(Rv)
        sol_s.append(amp[m])

# Multiply by 10 to account for 10X probe - see lab notebook for more information
for i in range(len(sol_s)):
    sol_s[i] = sol_s[i] * 10

# Plot bifurcation diagram
plt.figure(figsize = (9, 6))
plt.plot(rv_s, sol_s, 'r.', markersize = 0.5)
# plt.xlim(40,140)
# plt.ylim(-0.1, 1.1)
plt.xlim(45, 250)
plt.ylim(-0.1, 1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.title("Bifurcation Diagram")
plt.ylabel('X (V)', fontsize=20)
plt.xlabel('$R_v$ ($k\Omega$)', fontsize=20)
# plt.savefig("Bifurcation (exp), Rv250.pdf")
plt.show()
```

```python
#Compare experimental to theoretical (bifurcation diagram)

# Define model to solve ODE
def dx_dt(t, x):

    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    return (x2, x3, (-(R / Rv) * x3) - x2 - ((R2 / R1) * min(x1, 0)) - ((R / R0) * V0))

# Set initial parameter values
Rv_val = np.linspace(40, 140, 150)
R = 47
V0 = 0.250
R0 = 157
R2 = 6 
R1 = 1     
t_eval = np.linspace(700, 1200, 1000)
ts = [0, 1200]
rv = []
sol = []

# solve ODE
for i in Rv_val:
    Rv = i
    res = sp.integrate.solve_ivp(dx_dt, ts, y0 = [0.25, 0, 0], t_eval = t_eval, rtol = 1e-9)

    amp = []
    for i in range(len(res.y[0])):
        if (i > 0 and i < (len(res.y[0])-1)):
            prev = res.y[0][i-1]
            current = res.y[0][i]
            nextt = res.y[0][i+1]
            if (prev < current and nextt < current):
                amp.append(current)
    
    for i in range(len(amp)):
        rv.append(Rv)
        sol.append(amp[i])

# Plot theoretical in black and experimental in red
plt.figure(figsize = (9, 6))
plt.plot(rv, sol, 'k.', markersize = 0.5, label = 'Simulation')
plt.plot(rv_s, sol_s, 'r.', markersize = 0.5, label = 'Experimental Data')
# plt.title("Bifurcation Diagrams - Simulated and Experimental")
plt.ylabel('X (V)', fontsize =20)
plt.xlabel('$R_v$ ($k\Omega$)', fontsize=20)
plt.legend(loc="upper left", prop={"size": 13})
plt.xlim(45, 250)
plt.ylim(-0.1, 1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.savefig("Birfurcation Sim+exp, Rv250.pdf")
plt.show()
```

```python
# Plot region where we will take the percent difference of the rv values between experiment and simulation

plt.figure(figsize = (9, 6))
plt.plot(rv[0:10000], sol[0:10000], 'k.', markersize = 0.5, label = 'Simulation')
plt.plot(rv_s[0:35000], sol_s[0:35000], 'r.', markersize = 0.5, label = 'Experimental Data')
# plt.title("Bifurcation Diagrams - Simulated and Experimental")
plt.ylabel('X (V)', fontsize =20)
plt.xlabel('$R_v$ ($k\Omega$)', fontsize=20)
plt.legend(loc="upper left", prop={"size": 13})
# plt.xlim(45, 250)
plt.ylim(-0.1, 1.5)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.savefig("Birfurcation Sim+exp, Rv250.pdf")
plt.show()
```

```python
# find average percent difference of rv values between experimental and simulated

# choose up to Rv~130
n_rv, n_sol = rv[0:10000], sol[0:10000]
n_rv_s, n_sol_s = rv_s[0:35000], sol_s[0:35000]

# calculate percent difference
perc_diff = []
for i in range(len(n_sol)):
    j = i * 3 # decimate experimental data since number of exp data points is 3.5x number of simulated data points
    if (n_sol_s[j] >= n_sol[i] - 0.005 and n_sol_s[j] <= n_sol[i] + 0.005):
        percent_diff = (np.abs(n_rv[i]-n_rv_s[j])/n_rv[i])*100
        perc_diff.append(percent_diff)

sum_ = 0
for i in range(len(perc_diff)):
    sum_ = sum_ + perc_diff[i]

avg_perc_diff = sum_ / len(perc_diff)

print(f"Average percent difference between simulated and experimental Rv values: {avg_perc_diff:.1f}%")
```

### Calculation of Constants

```python
# feigenbaum constant = delta_n/delta_plus_1, where delta is the width between solutions (in kOhms)
```

```python
# Find values from graph 
delta_n = 69.5-55.7 # These were the values observed for bifurcation points in the "time series plots and phase diagrams" section
delta_n_err = np.sqrt((0.695**2) + (0.557**2)) # error in resistance values of variable resistor is 1%
```

```python
# find values from graph
delta_n_plus_1 = 72.5-69.5 # These were the values observed for bifurcation points in the "time series plots and phase diagrams" section
delta_n_plus_1_err = np.sqrt((0.725**2) + (0.695**2))
```

```python
# calculate correlated data
err_corr = np.sqrt(((delta_n_err / delta_n)**2) + ((delta_n_plus_1_err / delta_n_plus_1)**2) - ((2 / (delta_n * delta_n_plus_1)) * (((delta_n_err * delta_n_plus_1_err)**2) / (0.695**2))))
```

```python
# Finding constants

# Calculate feigenbaum constant and the percent diff from actual value
feigenbaum_act = 4.669
feigenbaum_est = (delta_n)/(delta_n_plus_1)
err_f = feigenbaum_est * err_corr
f_per_diff = (np.abs(feigenbaum_act-feigenbaum_est)/feigenbaum_act)*100
print(f"Estimated feigenbaum constant: {feigenbaum_est:.1f}  {err_f:.1f}")
print(f"Percent difference from actual value (4.669): {f_per_diff:.2f}%")
```

### Technical Report Plots

```python
# setup for simulations

# define constants 
R = 47
V0 = 0.250
R0 = 157
R2 = 6 
R1 = 1 
R2_R1= R2/R1
C = 1e-9
time_const = (R*1000)*C
ts = np.linspace(100, 200, 1000)

# define model to solve ODE
def dx_dt(t, x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    return (x2, x3, (-(R / Rv) * x3) - x2 - ((R2 / R1) * min(x1, 0)) - ((R / R0) * V0))
```

```python
# define fucntion to make plots for technical report

def tech_plot(t, dx, x, Rv, xtick, ytick, xtick2, ytick2, typ):
    import matplotlib.pyplot as plt
    plt.rcParams.update({'font.size': 16})
    # make size of plot 80% of 6.5" text width and with a 3:2 aspect ratio for the height
    w_text = 6.5
    w = 0.8 * w_text
    h = (2.0 / 3.0) * w

    # plot voltage
    fig, axs = plt.subplots(2, sharex=True, figsize = (w, h))
    # plt.ylabel('Voltage (V)', fontsize = 14)
    plt.xlabel('T (t/RC)', fontsize = 14)
    ylabel = 'Voltage (V)'
    vlabel1 = '$x$'
    vlabel2 = '$\dot{x}$'
    plt.figtext(0.03, 0.63, ylabel, wrap = True, horizontalalignment = 'center', fontsize = 14, rotation = 90)
    plt.figtext(0.93, 0.9, vlabel1, wrap = True, horizontalalignment = 'center', fontsize = 12)
    plt.figtext(0.93, 0.5, vlabel2, wrap = True, horizontalalignment = 'center', fontsize = 12)
    plt.xticks(xtick2)
    axs[0].plot(t, x*10, 'k', markersize = 0.5, label = 'x')
    axs[1].plot(t, dx, 'k', markersize = 0.5, label = '$\dot{x}$')
    plt.yticks(ytick2)
    plt.tight_layout()
    # plt.savefig('technical_r_figs/zr-V-Rv'+str(Rv)+'-'+str(typ)+'.png')
    plt.show()

    # plot phase diagram
    plt.figure(figsize = (w, h))
    plt.plot(x*10, dx, 'k.', markersize = 0.5)
    plt.ylabel('Voltage (V)', fontsize = 14)
    plt.xlabel('Voltage (V)', fontsize = 14)
    plt.yticks(ytick)
    plt.xticks(xtick)
    plt.tight_layout()
    # plt.savefig('technical_r_figs/zr-phase-Rv'+str(Rv)+'-'+str(typ)+'.png')
    plt.show()
```
